```python
import cv2
import numpy as np
import tensorflow
from tensorflow.keras.models import load_model
#Load the CNN Model
new_model = load_model('face_detector_model.h5')
# CNN model labels.
emotions = ["angry", "fear", "happy", "neutral", "sad", "surprise"]

# Load the Cascade Classifier
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def img_predict(path):
    source = cv2.imread(path)
    try:
            # Convert to grayscale
        gray_img = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)

            # Detect the faces
        faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, 
                        minNeighbors=3, 
                        minSize=(30,30))

        for (x, y, w, h) in faces:
                #display rectrangle
            cv2.rectangle(source, (x, y), (x+w, y+h), (0, 0, 255), 3)

                # roi_color =  on the original image for face image.
            roi = source[y:y+h, x:x+w]

                # source preprocessing for cnn model predictions.
            final = cv2.resize(roi, (48, 48))
            final = np.expand_dims(final, axis=0)
            final = final/255.0
                #predictions.
            prediction = new_model.predict(final)
            prediction_label = emotions[np.argmax(prediction)]

                #Predictions visualize.
            org = (x, y-10)
            fontFace = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (0, 0, 255)
            lineType = cv2.LINE_4
            img_text = cv2.putText(source, prediction_label, org, fontFace,
                                        fontScale, color, lineType)
    except:
        pass

    return source,prediction_label
```

```python
def real_predict():
    cap = cv2.VideoCapture(0)
    while True:
        _,source = cap.read()
        try:

            gray_img = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)

                # Detect the faces
            faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, 
                            minNeighbors=3, 
                            minSize=(30,30))

            for (x, y, w, h) in faces:
                    #display rectrangle
                cv2.rectangle(source, (x, y), (x+w, y+h), (0, 0, 255), 3)

                    # roi_color =  on the original image for face image.
                roi = source[y:y+h, x:x+w]

                    # source preprocessing for cnn model predictions.
                final = cv2.resize(roi, (48, 48))
                final = np.expand_dims(final, axis=0)
                final = final/255.0
                    #predictions.
                prediction = new_model.predict(final)
                prediction_label = emotions[np.argmax(prediction)]

                    #Predictions visualize.
                org = (x, y-10)
                fontFace = cv2.FONT_HERSHEY_SIMPLEX
                fontScale = 1
                color = (0, 0, 255)
                lineType = cv2.LINE_4
                img_text = cv2.putText(source, prediction_label, org, fontFace,
                                            fontScale, color, lineType)

        except:
            pass

        # display image
        cv2.imshow('Result', source)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
# Emotions Detections on the human face's

true_ = ['neutral','happy','neutral','angry','fear','angry','neutral','fear','surprise','happy','neutral','neutral','happy','sad','surprise','neutral','sad','happy','sad','neutral']
count = 0
for i in range(1,21):
    result,text = img_predict('pics/test/test0x0'+str(i)+'.jpg')
    if text == true_[i-1]:
        print('SUCCESFULLY!     ' + 'Predicted: ' + str(text) + ' ' + 'True Value:' + str(true_[i-1]))
        count = count + 1
    else:
        print('WRONG!     ' + 'Predicted: ' + str(text) + ' ' + 'True Value:' + str(true_[i-1]))
    while True:
        cv2.imshow('Result', result)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cv2.destroyAllWindows()
print('\n\nSuccesfully Rate: %' + str(count*5))
```
